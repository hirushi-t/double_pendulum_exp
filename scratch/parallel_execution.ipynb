# Threading

- A thread is subset of the process.
- Every time you open an application, you create a thread.
- Usually, a process has only 1 thread of control (i.e one set of machine instructions executing concurrently).
- However, a process may be made up of multiple threads executing in parallel.
- Threads can only truly run in parallel in a multi-core CPU. So, 1 CPU executes 1 process, and each core executes 1 thread. On single-core CPUs, the OS will rapidly switch between threads to execute them concurrently, as only one thread can execute an any given instant.


# Process

- This is an executing instance of a program - also known as a task.
- On a multiprocesser system (>1 CPU), multiple processes can be executed in parallel.

# Raspberry Pi 5

- Single processor system. 
- 4 separate cores inside its main processor.
- Can run up to 4 threads in parallel (truly).
- Can run more concurrently.
